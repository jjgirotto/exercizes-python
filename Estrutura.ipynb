### 3. Elabore uma estrutura para representar um produto (código, nome, preço). **Crie uma função** para cadastrar 5 produtos. **Crie outra função** para aplicar 10% de aumento no preço do produto e apresente, por meio **de outra função**, todos os dados do produtos cadastrados, após o aumento. Construa uma função para cada opção do menu a seguir:
Menu do Sistema
1. Cadastrar
2. Reajustar
3. Visualizar
4. Sair

Qual opção deseja?

```
class Produto:
    codigo = 0
    nome = ''
    preco = 0.0

def menu():
    print('\nMenu do sistema')
    print('1 - Cadastro de produtos')
    print('2 - Reajuste de produtos')
    print('3 - Visualização de produtos')
    print('4 - Sair')
    opcao = int(input('Digite a opção desejada: '))
    return opcao

def cadastrar(vet_produto):
    for i in range(5):
        p = Produto()
        p.codigo = int(input('Cadastre o código do produto: '))
        p.nome = input('Cadastre o nome do produto: ')
        p.preco = float(input('Cadastre o preço do produto: R$'))
        vet_produto.append(p)
    print('Cadastro realizado com sucesso!')
    return vet_produto

def visualizar(vet_produto):
    if len(vet_produto) > 0:
        print('Código   Nome    Preço')
        for i in range(len(vet_produto)):
            print(vet_produto[i].codigo, '\t', vet_produto[i].nome, '\t', vet_produto[i].preco)
    else:
        print('Não há produtos cadastrados!')

def reajustar(vet_produto):
    if len(vet_produto) > 0:
        for i in range(len(vet_produto)):
            vet_produto[i].preco *= 1.1
        print('Reajuste realizado com sucesso!')
    else:
        print('Não há produtos cadastrados!')
    return vet_produto

def main():
    vet_produto = []
    op = menu()
    while op >= 1 and op <= 3:
        if op == 1:
            vet_produto = cadastrar(vet_produto)
        elif op == 2:
            vet_produto = reajustar(vet_produto)
        elif op == 3:
            visualizar(vet_produto)
        op = menu()

main()
```

### 4. Uma escola precisa montar o cadastro geral de seus alunos. Este cadastro deverá conter as seguintes informações por aluno: nome completo, data de nascimento, telefone, endereço e série atual. Levando em conta que esta escola possui no máximo 500 alunos, como você faria para estruturar estas informações num sistema de gerenciamento para a escola? Implemente utilizando estrutura. Também **use a criação de funções para cada operação**. Use o menu a seguir:
Menu de opções:
1. Cadastrar alunos
2. Consulta por nome
3. Visualizar todos os dados
4. Sair


```
from datetime import datetime

class Aluno:
    nome = ''
    data_nasc = ''
    telefone = 0
    endereco = ''
    serie = 0

def menu():
    print('\nMenu do sistema')
    print('1 - Cadastro de alunos\n2 - Consulta de alunos por nome\n3 - Visualização de dados\n4 - Sair')
    opcao = int(input('Digite a opção desejada: '))
    return opcao

def cadastrar(vet_alunos):
    for i in range(10):
        a = Aluno()
        a.nome = input('Cadastre o nome do aluno: ')
        a.data_nasc = input('Cadastre a data de nascimento: (dd/mm/aaaa) ')
        a.data_nasc = datetime.strptime(a.data_nasc, "%d/%m/%Y")
        a.telefone = int(input('Cadastre o número de telefone (sem hífens): '))
        a.endereco = input('Cadastre o endereço completo: ')
        a.serie = int(input('Cadastre a série atual: '))
        vet_alunos.append(a)
    print('Cadastro realizado com sucesso!')
    return vet_alunos

def consultar(vet_alunos):
    if len(vet_alunos) > 0:
        encontrado = False
        nome = input('Digite o nome para consulta: ')
        for i in range(len(vet_alunos)):
            if vet_alunos[i].nome == nome:
                encontrado = True
                break
        if encontrado:
            print('Nome\tData de nascimento\tTelefone\tEndereço\t\t\t\t\t\tSérie')
            print(f'{vet_alunos[i].nome}\t{vet_alunos[i].data_nasc.strftime("%d/%m/%Y")}\t\t{vet_alunos[i].telefone}\t{vet_alunos[i].endereco}\t{vet_alunos[i].serie}')
        else:
            print('Aluno não encontrado!')
    else:
        print('Não há alunos cadastrados!')

def visualizar(vet_alunos):
    if len(vet_alunos) > 0:
        print('Nome\tData de nascimento\tTelefone\tEndereço\t\t\t\t\t\tSérie')
        for i in range(len(vet_alunos)):
            print(f'{vet_alunos[i].nome}\t{vet_alunos[i].data_nasc.strftime("%d/%m/%Y")}\t\t{vet_alunos[i].telefone}\t{vet_alunos[i].endereco}\t{vet_alunos[i].serie}')
    else:
        print('Não há alunos cadastrados!')

def main():
    vet_alunos = []
    op = menu()
    while op >= 1 and op <= 3:
        if op == 1:
            vet_alunos = cadastrar(vet_alunos)
        elif op == 2:
            consultar(vet_alunos)
        elif op == 3:
            visualizar(vet_alunos)
        op = menu()

main()
```

### 5. Faça um programa que realize o cadastro de contas bancarias com as seguintes informações: numero da conta, nome do titular e saldo. O banco permitirá o cadastramento de 15 contas. **Crie uma função para cada opção do menu** a seguir. Utilize a estrutura na passagem de parâmetro:
Menu de opções:
1. Cadastrar contas
2. Visualizar todas as contas
3. Consultar por nome
4. Alterar nome e/ou saldo de um número de conta
5. Excluir a conta com menor saldo
6. Sair

Observação:

* No item de menu 4. Alterar nome e/ou saldo de um número de conta, entenda que apenas pode ser alterado o nome e o saldo depois que você pesquisar pelo número da conta.
* No item 5 do menu, encontre o menor saldo dentre todos cadastrados, **depois exclua esta ÚNICA conta.**. O assunto de excluir algo de um vetor foi dado na disciplina de Algoritmo.



```
class ContaBancaria:
    numero = 0
    nome = ''
    saldo = 0.0

def menu():
    print('\nMenu do opções')
    print('1 - Cadastro de contas\n2 - Visualização de contas\n3 - Consulta por nome\n4 - Alterar nome/saldo na conta\n5 - Excluir a conta com menor saldo\n6 - Sair')
    opcao = int(input('Digite a opção desejada: '))
    return opcao

def cadastrar(vet_contas):
    for i in range(15):
        c = ContaBancaria()
        c.numero = int(input('Cadastre o número da conta: '))
        c.nome = input('Cadastre o nome: ')
        c.saldo = float(input('Cadastre o valor do saldo: R$ '))
        vet_contas.append(c)
    print('Cadastro realizado com sucesso!')
    return vet_contas

def consultar(vet_contas):
    if len(vet_contas) > 0:
        encontrado = False
        nome = input('Digite o nome para consulta: ')
        for i in range(len(vet_contas)):
            if vet_contas[i].nome == nome:
                encontrado = True
                print('Número\tNome\tSaldo')
                print(f'{vet_contas[i].numero}\t{vet_contas[i].nome}\t{vet_contas[i].saldo}')
                break
        if not encontrado:
            print('Conta não encontrada!')
    else:
        print('Não há contas cadastradas!')


def visualizar(vet_contas):
    if len(vet_contas) > 0:
        print('Número\tNome\tSaldo')
        for i in range(len(vet_contas)):
            print(f'{vet_contas[i].numero}\t{vet_contas[i].nome}\t{vet_contas[i].saldo}')
    else:
        print('Não há contas cadastradas!')

def alterar(vet_contas):
    if len(vet_contas) > 0:
        numero = int(input('Digite o número da conta: '))
        for i in range(len(vet_contas)):
            if numero == vet_contas[i].numero:
                op = int(input('1. Alterar nome da conta\n2. Alterar saldo da conta.\nDigite a opção desejada: '))
                if op == 1:
                    vet_contas[i].nome = input('Digite um novo nome para alteração: ')
                elif op == 2:
                    vet_contas[i].saldo = float(input('Digite um novo saldo para alteração: '))
        print('Alteração realizada com sucesso!')
    else:
        print('Não há contas cadastradas!')
    return vet_contas

def excluir(vet_contas):
    if len(vet_contas) > 0:
        for i in range(len(vet_contas)):
            if i == 0:
                menor_saldo = vet_contas[i].saldo
                indice_menor = i
            if vet_contas[i].saldo < menor_saldo:
                menor_saldo = vet_contas[i].saldo
                indice_menor = i
        vet_contas.pop(indice_menor)
        print('Exclusão realizada com sucesso!')
    else:
        print('Não há contas cadastradas!')
    return vet_contas

def main():
    vet_contas = []
    op = menu()
    while op >= 1 and op <= 5:
        if op == 1:
            vet_contas = cadastrar(vet_contas)
        elif op == 2:
            visualizar(vet_contas)
        elif op == 3:
            consultar(vet_contas)
        elif op == 4:
            vet_contas = alterar(vet_contas)
        elif op == 5:
            vet_contas = excluir(vet_contas)
        op = menu()

main()
```

```
class Cliente:
    cod_cli = 0
    nome = ''
    fone = 0

class Documento:
    num_doc = 0
    dia_venc = 0
    cod_cli = 0
    dia_pag = 0
    valor = 0.0
    juros = 0.0

def menu():
    print('\nMenu do sistema\n')
    print('1-Cadastrar clientes\n2-Relatório de clientes\n3-Cadastrar documentos\n4-Relatório de documentos')
    print('5-Excluir clientes sem documentos\n6-Excluir documentos individuais pelo número\n7-Excluir documentos por cliente')
    print('8-Excluir documentos por período\n9-Alterar as informações dos clientes\n10-Mostrar o total de documentos de determinado cliente')
    print('11-Sair.\n')
    opcao = int(input('Digite a opção desejada: '))
    return opcao

def cadastrar_clientes(vet_cliente):
    if len(vet_cliente) < 15:
        cod_cli = int(input('Cadastre o código do cliente: '))
        codigo_igual = False
        for i in range(len(vet_cliente)):
            if cod_cli == vet_cliente[i].cod_cli:
                #print('Este código já foi cadastrado, insira um código novo.')
                codigo_igual = True
                break
        if not codigo_igual: #perguntando se o código é falso!!!!
                c = Cliente()
                c.cod_cli = cod_cli
                c.nome = input('Cadastre o nome do cliente: ')
                c.fone = int(input('Cadastre o telefone do cliente: '))
                vet_cliente.append(c)
                print('Cadastro realizado com sucesso!')
        elif codigo_igual:
            print('Este código já foi cadastrado, insira um código novo.')

    else:
        print('O limite de 15 clientes cadastrados foi atingido.')

    return vet_cliente

def listar_cliente(vet_cliente):
    if len(vet_cliente) > 0:
        print('Código\tNome\tTelefone')
        for i in range(len(vet_cliente)):
            print(f'{vet_cliente[i].cod_cli}\t{vet_cliente[i].nome}\t{vet_cliente[i].fone}')
    else:
        print('Não há clientes cadastrados!')

def cadastrar_docs(vet_cliente, vet_docs):
    if len(vet_docs) < 30:
        cod_cli = int(input('Insira o código do cliente: '))
        codigo_existente = False
        for i in range(len(vet_cliente)):
            if cod_cli == vet_cliente[i].cod_cli:
                codigo_existente = True
                d = Documento()
                d.cod_cli = cod_cli
                d.num_doc = int(input('Cadastre o número do documento: '))
                d.dia_venc = int(input('Cadastre o dia do vencimento: '))
                d.dia_pag = int(input('Cadastre o dia do pagamento: '))
                d.valor = float(input('Cadastre o valor do documento: '))
                if d.dia_pag > d.dia_venc:
                    d.juros = d.valor * 0.05
                else:
                    d.juros = 0
                vet_docs.append(d)
                print('Cadastro realizado com sucesso!')
        if not codigo_existente:
            print('Não há cliente cadastrado com este código.')
    else:
        print('O limite de 30 documentos cadastrados foi atingido.')
    return vet_docs

def listar_docs(vet_docs):
    if len(vet_docs) > 0:
        print('Número\tDia vencimento\tCód. Cliente\tDia pagamento\tValor R$\tJuros R$')
        for i in range(len(vet_docs)):
            print(f'{vet_docs[i].num_doc}\t\t{vet_docs[i].dia_venc}\t\t{vet_docs[i].cod_cli}\t\t{vet_docs[i].dia_pag}\t{vet_docs[i].valor:.2f}\t\t{vet_docs[i].juros:.2f}')
    else:
        print('Não há documentos cadastrados!')

def excluir_cliente_sem_docs(vet_cliente, vet_docs):
    if len(vet_cliente) > 0 and len(vet_docs) > 0:
        doc_encontrado = False
        cli_encontrado = False
        codigo = int(input('Digite o código do cliente a ser excluído: '))
        contador = 0
        for c in range(len(vet_cliente)):
            if codigo == vet_cliente[c].cod_cli:
                cli_encontrado = True
                for i in range(len(vet_docs)):
                    if codigo == vet_docs[i].cod_cli:
                        doc_encontrado = True
                        break
                    else:
                        contador += 1
                        if contador == len(vet_docs):
                            indice = i
        if doc_encontrado:
            print('Este cliente possui documentos e não pode ser excluído.')
        if not cli_encontrado:
            print('Cliente não encontrado.')
        if cli_encontrado and not doc_encontrado:
            vet_cliente.pop(indice)
            print('Cliente excluído com sucesso.')
    else:
        print('Para excluir clientes sem documentos, deve haver clientes e documentos cadastrados.')
    return vet_cliente

def excluir_docs_num(vet_docs):
    if len(vet_docs) > 0:
        cont = 0
        numero = int(input('Digite o número do documento a excluir: '))
        for i in range(len(vet_docs)):
            if numero == vet_docs[i].num_doc:
                indice = i
                break
            else:
                cont += 1
        if cont == len(vet_docs):
            print('Documento não encontrado.')
        else:
            vet_docs.pop(indice)
            print('Documento excluído com sucesso.')
    else:
        print('Não há documentos cadastrados!')
    return vet_docs

def excluir_docs_cliente(vet_cliente, vet_docs):
    if len(vet_cliente) > 0 and len(vet_docs) > 0:
        doc_encontrado = False
        cli_encontrado = False
        codigo = int(input('Insira o código do cliente que deseja excluir os documentos: '))
        for c in range(len(vet_cliente)):
            if codigo == vet_cliente[c].cod_cli:
                cli_encontrado = True
                indices = []
                for i in range(len(vet_docs)):
                    if codigo == vet_docs[i].cod_cli:
                        doc_encontrado = True
                        indices.append(i)
        if doc_encontrado:
            ind = len(indices) - 1
            for i in range(len(indices)):
                vet_docs.pop(indices[ind])
                ind -= 1
            print(f'Houve {len(indices)} documento(s) excluídos(s) com sucesso.')
        if not cli_encontrado:
            print('Cliente não encontrado.')
        if cli_encontrado and not doc_encontrado:
            print('Este cliente não possui documentos.')
    else:
        print('Para mostrar documentos de determinado cliente, deve haver clientes e documentos cadastrados.')
    return vet_docs

def excluir_docs_periodo(vet_docs):
    if len(vet_docs) > 0:
        dia_inicial = int(input('Informe o dia inicial: '))
        dia_final = int(input('Informe o dia final: '))
        indices = []
        encontrados = False
        for i in range(len(vet_docs)):
            if vet_docs[i].dia_venc >= dia_inicial and vet_docs[i].dia_venc <= dia_final:
                indices.append(i)
                encontrados = True
        if encontrados:
            ind = len(indices) - 1
            for i in range(len(indices)):
                vet_docs.pop(indices[ind])
                ind -= 1
            print(f'Houve {len(indices)} documento(s) excluídos(s) com sucesso.')
        else:
            print('Não há documentos a vencer neste período informado.')
    else:
        print('Não há documentos cadastrados!')
    return vet_docs

def alterar(vet_cliente):
    if len(vet_cliente) > 0:
        codigo = int(input('Digite o código do cliente para alterar os dados: '))
        encontrado = False
        for i in range(len(vet_cliente)):
            if codigo == vet_cliente[i].cod_cli:
                vet_cliente[i].nome = input('Digite o novo nome: ')
                vet_cliente[i].fone = int(input('Digite o novo telefone: '))
                encontrado = True
                print('Alterações realizadas com sucesso!')
                break
        if not encontrado:
            print('Cliente não encontrado. Digite um código de cliente válido!')
    else:
        print('Não há clientes cadastrados!')
    return vet_cliente

def mostrar_docs_cliente(vet_cliente, vet_docs):
    if len(vet_cliente) > 0 and len(vet_docs) > 0:
        doc_encontrado = False
        cli_encontrado = False
        codigo = int(input('Insira o código do cliente: '))
        for c in range(len(vet_cliente)):
            if codigo == vet_cliente[c].cod_cli:
                cli_encontrado = True
                indices = []
                for i in range(len(vet_docs)):
                    if codigo == vet_docs[i].cod_cli:
                        doc_encontrado = True
                        indices.append(i)
        if doc_encontrado:
            print('Númeroz\tDia vencimento\tCód. Cliente\tDia pagamento\tValor R$\tJuros R$')
            for i in range(len(indices)):
                print(f'{vet_docs[indices[i]].num_doc}\t\t{vet_docs[indices[i]].dia_venc}\t\t{vet_docs[indices[i]].cod_cli}\t\t{vet_docs[indices[i]].dia_pag}\t{vet_docs[indices[i]].valor:.2f}\t\t{vet_docs[indices[i]].juros:.2f}')
        if not cli_encontrado:
            print('Este cliente não existe!')
        if cli_encontrado and not doc_encontrado:
            print('Este cliente não possui documentos.')
    else:
        print('Para mostrar documentos de determinado cliente, deve haver clientes e documentos cadastrados.')

def main():
    vet_cliente = []
    vet_docs = []
    op = menu()
    while op >= 1 and op <= 10:
        if op == 1:
            vet_cliente = cadastrar_clientes(vet_cliente)
        elif op == 2:
            listar_cliente(vet_cliente)
        elif op == 3:
            vet_docs = cadastrar_docs(vet_cliente, vet_docs)
        elif op == 4:
            listar_docs(vet_docs)
        elif op == 5:
            vet_cliente = excluir_cliente_sem_docs(vet_cliente, vet_docs)
        elif op == 6:
            vet_docs = excluir_docs_num(vet_docs)
        elif op == 7:
            vet_docs = excluir_docs_cliente(vet_cliente, vet_docs)
        elif op == 8:
            vet_docs = excluir_docs_periodo(vet_docs)
        elif op == 9:
            vet_cliente = alterar(vet_cliente)
        elif op == 10:
            mostrar_docs_cliente(vet_cliente, vet_docs)
        op = menu()

main()
```

####9. (OPCIONAL A IMPLEMENTAÇÃO) Uma empresa prestadora de serviços armazena informações sobre os serviços prestados. Sabe-se que a empresa pode realizar no máximo três serviços diariamente. É de interesse de sua direção manter um histórico mensal (30 dias) sobre os serviços prestados.
A empresa realiza quatro tipos de serviços:

1) pintura;

2) jardinagem;

3) faxina e

4) reforma em geral.

Cada serviço realizado deve ser cadastrado com as seguintes informações: número do serviço, valor do serviço, código do serviço e código do cliente, **numa matriz 30x3 referente a estrutura Servico_prestado**.

Cadastre/digite os quatro tipos de serviços (código e descrição) que a empresa poderá realizar. Para isso, utilize um vetor de quatro posições referente a **estrutura Tipo_servico**.
O programa deverá mostrar o seguinte menu de opções:
1. Cadastrar os tipos de serviços
2. Mostrar todos os tipos de serviço
3. Cadastrar os serviços prestados
4. Mostrar todos os serviços prestados
5. Mostrar os serviços prestados em determinado dia
6. Mostrar os serviços prestados dentro de um intervalo de valor
7. Mostrar um relatório geral (separado por dia), que exiba, inclusive, a descrição do tipo do serviço
8. Sair

**Para a opção 1**: deve-se cadastrar os tipos de serviços oferecidos pela empresa, com código e descrição.

**Para a opção 3**: deve-se considerar que deverão ser cadastrados os serviços prestados ao logo do mês. Em cada dia podem ser cadastrados, no máximo, três serviços prestados.

Utilize uma matriz capaz de armazenar em cada posição todas as informações referentes a um serviço prestado (número, valor, código do serviço, código do cliente). **Cada linha representa um dia do mês**. Dessa maneira, considere a matriz com dimensão 30 × 3.

Solicite o dia em que o serviço foi prestado e as demais informações.
Lembre-se de que a empresa só pode prestar os serviços que já tenham sido cadastrados no vetor de tipo de serviços.

Caso o usuário digite um código de tipo de serviço inválido, o programa deverá mostrar uma mensagem de erro.

Quando o usuário tentar cadastrar mais de três serviços prestados em um mesmo dia, também deverá mostrar uma mensagem de erro.

**Para a opção 5**: o programa deverá receber o dia que se deseja consultar e mostrar os respectivos serviços prestados.

**Para a opção 6**: o programa deverá receber o valor mínimo e o valor máximo e mostrar os serviços prestados que estiverem nesse intervalo.

**Para a opção 7**: o programa deverá mostrar todos os serviços prestados, conforme o exemplo a seguir.

```
class Servico_Prestado:
    n = 0
    valor = 0.0
    cod_serv_prestado = 0
    cod_cli = 0

class Tipo_servico:
    cod_serv = 0
    descricao = ''

def menu():
    print('\nMenu do sistema\n')
    print('1-Cadastrar os tipos de serviços\n2-Mostrar os tipos de serviços\n3-Cadastrar serviços prestados\n4-Mostrar serviços prestados')
    print('5-Mostrar os serviços prestados em determinado dia\n6-Mostrar os serviços prestados dentro de um intervalo de valor')
    print('7-Mostrar um relatório geral (separado por dia), que exiba, inclusive, a descrição do tipo do serviço\n8-Sair.')
    opcao = int(input('Digite a opção desejada: '))
    return opcao

def cadastrar_tipos(tipos):
    servicos = ['Pintura', 'Jardinagem', 'Faxina', 'Reforma geral']
    for i in range(4):
        tipo = Tipo_servico()
        tipo.cod_serv = i+1
        tipo.descricao = servicos[i]
        tipos.append(tipo)
        print(f'Cadastro de {tipos[i].descricao} realizado.')
    print('Cadastros realizados.')
    return tipos

def listar_tipos(tipos):
    if len(tipos) > 0:
        print('\nCódigo do serviço   Descrição')
        for i in range(len(tipos)):
            print(f'       {tipos[i].cod_serv}            {tipos[i].descricao}')
    else:
        print('Não há tipos de serviços cadastrados!')

def cadastrar_matriz():
    matriz = []
    for i in range(30):
        vetor = []
        for j in range(3):
            prestado = Servico_Prestado()
            prestado.n = 0
            prestado.valor = 0.0
            prestado.cod_serv_prestado = 0
            prestado.cod_cli = 0
            vetor.append(prestado)
        matriz.append(vetor)
    return matriz

def cadastrar_serv_prestados(tipos, prestados):
    codigo = int(input('Digite o código do serviço prestado: '))
    dia = int(input('Digite o dia dos serviços prestados: '))
    contador = 0
    for i in range(len(tipos)):
        if codigo == tipos[i].cod_serv:
            for i in range(4):
                linha = dia - 1
                if i == 3:
                    print('Em cada dia podem ser cadastrados, no máximo, três serviços prestados.')
                    break
                if prestados[linha][i].n == 0:
                    prestados[linha][i].n = int(input('Cadastre o número do serviço prestado: '))
                    prestados[linha][i].valor = float(input('Cadastre o valor do serviço prestado: '))
                    prestados[linha][i].cod_serv_prestado = codigo
                    prestados[linha][i].cod_cli = int(input('Cadastre o código do cliente: '))
                    break
        else:
            contador += 1
    if contador == len(tipos):
        print('Este código de tipo de serviço não existe.')
    return prestados

def listar_serv_prestados(prestados):
    if len(prestados) > 0:
        print('\n......Serviços prestados......')
        for linha in range(len(prestados)):
            print(f'Dia {linha+1}:')
            print('Número\tValor R$\tCódigo serviço\t\tCódigo cliente')
            for coluna in range(len(prestados[0])):
                print(f'{prestados[linha][coluna].n}\t{prestados[linha][coluna].valor}\t\t\t{prestados[linha][coluna].cod_serv_prestado}\t\t\t{prestados[linha][coluna].cod_cli}')
    else:
        print('\nNão existe serviços prestados cadastrados!')

def relatorio(tipos, prestados):
    if len(tipos) > 0 and len(prestados) > 0:
        print('\n......Serviços prestados......')
        for linha in range(len(prestados)):
            print(f'Dia {linha+1}:')
            print('Número\t\tValor R$\tCódigo serviço\t\tDescrição\tCódigo cliente')
            for coluna in range(len(prestados[0])):
                cod_descricao = prestados[linha][coluna].cod_serv_prestado
                for i in range(len(tipos)):
                    if tipos[i].cod_serv == cod_descricao:
                        descricao = tipos[i].descricao
                        break
                    else:
                        descricao = ''
                print(f'{prestados[linha][coluna].n}\t\t{prestados[linha][coluna].valor}\t\t\t{prestados[linha][coluna].cod_serv_prestado}\t\t{descricao}\t\t{prestados[linha][coluna].cod_cli}')
    else:
        print('\nNão existe serviços prestados cadastrados!')

def listar_serv_prestados_dia(prestados):
    if len(prestados) > 0:
        dia = int(input('Digite o dia que deseja consultar: '))
        print('\n......Serviços prestados......')
        print('Número\tValor R$\tCódigo serviço\t\tCódigo cliente')
        linha = dia - 1
        for coluna in range(len(prestados[dia-1])):
            print(f'{prestados[linha][coluna].n}\t{prestados[linha][coluna].valor}\t\t\t{prestados[linha][coluna].cod_serv_prestado}\t\t\t{prestados[linha][coluna].cod_cli}')
    else:
        print('\nNão existe serviços prestados cadastrados!')

def listar_serv_prestados_intervalo_valor(prestados):
    if len(prestados) > 0:
        dia_inicial = int(input('Digite o dia inicial do intervalo que deseja consultar: '))
        dia_final = int(input('Digite o dia inicial do intervalo que deseja consultar: '))
        print('\n......Serviços prestados......')
        print('Número\tValor R$\tCódigo serviço\t\tCódigo cliente')
        for linha in range(dia_inicial-1, dia_final):
            print(f'Dia {linha+1}:')
            for coluna in range(len(prestados[linha])):
                print(f'{prestados[linha][coluna].n}\t{prestados[linha][coluna].valor}\t\t\t{prestados[linha][coluna].cod_serv_prestado}\t\t\t{prestados[linha][coluna].cod_cli}')
    else:
        print('\nNão existe serviços prestados cadastrados!')

def main():
    op = menu()
    tipos = []
    serv_prestados = []
    while op >= 1 and op <= 7:
        if op == 1:
            tipos = cadastrar_tipos(tipos)
        elif op == 2:
            listar_tipos(tipos)
        elif op == 3:
            if len(tipos) > 0 and len(serv_prestados) == 0:
                serv_prestados = cadastrar_matriz()
            if len(tipos) > 0 and len(serv_prestados) > 0:
                serv_prestados = cadastrar_serv_prestados(tipos, serv_prestados)
            else:
                print('Cadastre os tipos de serviços.')
        elif op == 4:
            listar_serv_prestados(serv_prestados)
        elif op == 5:
            listar_serv_prestados_dia(serv_prestados)
        elif op == 6:
            listar_serv_prestados_intervalo_valor(serv_prestados)
        elif op == 7:
            relatorio(tipos, serv_prestados)
        op = menu()

main()
```
